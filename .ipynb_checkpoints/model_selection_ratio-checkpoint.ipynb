```python
import lightgbm as lgb
import xgboost as xgb
import catboost as ctb
```

```python
from sklearn import preprocessing
from sklearn.metrics import *
```

# load features

```python
feature_name = np.load(
    os.path.join(
    data_path,
    'feature_name.npy',
    ),
)

categorical_feature = np.load(
    os.path.join(
    data_path,
    'categorical_feature.npy',
    ),
)

numemrical_feature = np.load(
    os.path.join(
    data_path,
    'numemrical_feature.npy',
    ),
)
```

# load data

```python
data_filled = pd.read_parquet(
    os.path.join(
    data_path,
    'data_filled.parquet',
    )
)
```

```python
X = data_filled[feature_name]
```

```python
y_reg = data_filled['label__close_to_open_ratio']

y_cla = data_filled['label__rise']
```

# catboost model

## ratio

```python
%%time  

train_pool = ctb.Pool(
    X, 
    y_reg, 
    cat_features = categorical_feature,
    )

params = {
    'loss_function': 'MAE',
    'task_type':"GPU",
    'devices':'0',
    'iterations':1024,
    }

ctb_model = ctb.CatBoostRegressor(**params)

ctb_model.fit(train_pool)

ctb_model.save_model(
    fname = os.path.join(
    data_path,
    "daily_close_to_open_ratio.catboost"),
    )
```

```python
y_pred = ctb_model.predict(X)

median_train_ctb = median_absolute_error(
    y_reg,
    y_pred,
)

avg_train_ctb = mean_absolute_error(
    y_reg,
    y_pred,
)

print(f"""
model:\tCatBoost
task:\tClose to Open Ratio
median error - training:\t{median_train_ctb:0.4f}
average error - test:\t{avg_train_ctb:0.4f}
""")
```

## rise

```python
%%time  

train_pool = ctb.Pool(
    X, 
    y_cla, 
    cat_features = categorical_feature,
    )

params = {
    'loss_function': 'CrossEntropy',
    'task_type':"GPU",
    'devices':'0',
    'iterations':1024,
    }

ctb_model = ctb.CatBoostClassifier(**params)

ctb_model.fit(train_pool)

ctb_model.save_model(
    fname = os.path.join(
    data_path,
    "daily_rise.catboost"),
    )
```

```python
y_pred = ctb_model.predict_proba(X)[:,1]

fpr, tpr, thresholds = roc_curve(
    y_cla, y_pred, pos_label=1)
auc_ctb = auc(fpr, tpr)

average_precision_ctb = average_precision_score(
    y_cla,
    y_pred
    )

print(f"""
model:\tCatBoost
task:\tRise
AUC:\t{auc_ctb:0.4f},
Average Precision:\t{average_precision_ctb:0.4f}
""")
```

## feature importance

```python
feature_importance = []

for n, i in zip(ctb_model.feature_names_, ctb_model.feature_importances_):
    feature_importance.append({
        'feature_name':n,
        'feature_importance':i
    })
    
pd.DataFrame(feature_importance).to_sql(
    "feature_importance_catboost_rise", 
    conn,
    if_exists = "replace",
    )
```
