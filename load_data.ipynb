# increase

```python
qqq_weekly_ratio = sql(f"""
select 
a.Date,
a.Open,
a.Close,
a.Close/a.Open as daily_raito,
a.Close > a.Open as daily_increase,
b.Date as pre_date,
b.Close as pre_close,
a.Close/b.Close as weekly_ratio,
a.Close > b.Close as weekly_increase,
strftime('%w',a.Date) as current_week_day,
strftime('%w',b.Date) as pre_week_day
from qqq as a
left join qqq as b
on DATE(b.Date,'+7 days') = DATE(a.Date)
""")


.to_sql(
    'qqq_weekly_ratio', 
    conn,
    if_exists = "replace",
    )
```

```python
sql(f"""
select 
daily_increase,
count(*)
from qqq_weekly_ratio
group by daily_increase
""")
```

```python
sql(f"""
select 
weekly_increase,
count(*)
from qqq_weekly_ratio
group by weekly_increase
""")
```
