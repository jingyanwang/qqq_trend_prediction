```python
sql(f"""
select 
a.*,
b.price as price_ten_years_ago,
b.date as date_ten_years_ago,
a.price/b.price - 1 as return,
a.price/b.price as ratio
from qqq_montly as a
left join qqq_montly as b
on DATE(b.date,'1 years') = DATE(a.date)
""").to_sql(
    "qqq_with_ten_years_ago", 
    conn,
    if_exists = "replace",
    )
```

# plot

```python
df = sql(f"""
select 
date,
price,
'current' as type
from qqq_with_ten_years_ago
union 
select 
date,
price_ten_years_ago as price,
'1-year ago' as type
from qqq_with_ten_years_ago
""")
```

```python
fig = px.line(
    df, x="date", y="price", color='type', 
    )
fig.show()
```
